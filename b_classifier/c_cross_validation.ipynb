```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd

iris = load_iris()
features = iris.data
targets = iris.target

target_df = pd.DataFrame(targets, columns=['target'])
target_df.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold

dtc = DecisionTreeClassifier(random_state=124, min_samples_leaf=6)
kfold = KFold(n_splits=5)
```

```python
features.shape
```

```python
count = 0
for train_index, test_index in kfold.split(features):
    # ë¶„ë¦¬
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = targets[train_index], targets[test_index]
    
    # í•™ìŠµ ë° ì˜ˆì¸¡
    dtc.fit(X_train, y_train)
    prediction = dtc.predict(X_test)

    # í‰ê°€
    accuracy = np.round(accuracy_score(y_test, prediction), 4)

    # ê²€ì¦
    train_targets = pd.DataFrame(y_train)
    test_targets = pd.DataFrame(y_test)

    count += 1

    print(f'{count} íšŒì°¨')
    print(f'í•™ìŠµ íƒ€ê²Ÿ ë°ì´í„° ë¶„í¬: \n{train_targets.value_counts()}')
    print(f'ê²€ì¦ íƒ€ê²Ÿ ë°ì´í„° ë¶„í¬: {test_targets.value_counts()}')
    print(f'ì •í™•ë„: {accuracy}')
```

```python
from sklearn.model_selection import StratifiedKFold

s_kfold = StratifiedKFold(n_splits=5)
```

```python
count = 0
accuracy_list = []
# ë¹„ì¤‘ì„ ë§ì¶”ê¸° ìœ„í•´ íƒ€ê²Ÿë„ ì „ë‹¬ í•´ì•¼í•œë‹¤. K-Foldì™€ ë‹¤ë¥¸ ì !
for train_index, test_index in s_kfold.split(features, targets):
    # ë¶„ë¦¬
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = targets[train_index], targets[test_index]
    
    # í•™ìŠµ ë° ì˜ˆì¸¡
    dtc.fit(X_train, y_train)
    prediction = dtc.predict(X_test)

    # í‰ê°€
    accuracy = np.round(accuracy_score(y_test, prediction), 4)
    accuracy_list.append(accuracy)

    # ê²€ì¦
    train_targets = pd.DataFrame(y_train)
    test_targets = pd.DataFrame(y_test)

    count += 1

    print(f'{count} íšŒì°¨')
    print(f'í•™ìŠµ íƒ€ê²Ÿ ë°ì´í„° ë¶„í¬: \n{train_targets.value_counts()}')
    print(f'ê²€ì¦ íƒ€ê²Ÿ ë°ì´í„° ë¶„í¬: {test_targets.value_counts()}')
    print(f'ì •í™•ë„: {accuracy}')

print(f'í‰ê·  ì •í™•ë„: {accuracy_list}')
```

#### í¸í•˜ê²Œ ìˆ˜í–‰í•  ìˆ˜ ìˆëŠ” êµì°¨ ê²€ì¦
estimator = ëª¨ë¸ ê°ì²´, x= feature, y=target, scoring =accuracy ë¬¸ìì—´  
ë¶„ë¥˜í˜•ì—ì„œëŠ” íƒ€ê²Ÿì´ ë²”ì£¼í˜• starified fold, ìˆ˜ì¹˜í˜•ì—ì„œëŠ” k-fold ì‚¬ìš©  

**cross_val_score(estimator, x, y, cv, scoring)**
- estimator: classifier ì¢…ë¥˜ ëª¨ë¸ì´ë©´ ë‚´ë¶€ì ìœ¼ë¡œ starified K-Foldë¡œ ì§„í–‰ëœë‹¤.
- x: features
- y: targets
- cv: í´ë“œ ì„¸íŠ¸ ê°œìˆ˜
- scoring: í‰ê°€ í•¨ìˆ˜, ì •í™•ë„(accuracy)ì™¸ì— ë‹¤ë¥¸ ê²ƒì€ ë‹¤ë¥¸ ì¥ì—ì„œ ë°°ìš´ë‹¤.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
dtc = DecisionTreeClassifier(random_state=124, min_samples_leaf=6)

features = iris.data
targets = iris.target

score = cross_val_score(dtc, features, targets, cv=5, scoring='accuracy')
print(np.round(np.mean(score) , 4))
```

#### **GridSearchTV**
**GridSearchTV(estimator, param_grid, cv, refit, return_train_score)**
- estimator: í•™ìŠµí•  ëª¨ë¸ ê°ì²´ ì‘ì„±
- param_grid: dictí˜•íƒœë¡œ ì „ë‹¬í•´ì•¼ í•˜ë©°, ì£¼ìš” keyê°’ì€ max_depth, min_samples_splitì´ë‹¤.
- cv: í´ë“œ ì„¸íŠ¸ ê°œìˆ˜
- refit: ì „ë‹¬í•œ ëª¨ë¸ ê°ì²´ì— ìµœì ì˜ í•˜ì´í¼ íŒŒë¼ë¯¸í„°ë¡œ ì ìš©í•˜ê³ ì í•  ë•Œ
- return_train_score: êµì°¨ ê²€ì¦ ì ìˆ˜ë¥¼ ê°€ì ¸ì˜¬ ì§€ì— ëŒ€í•´ True ë˜ëŠ” Falseë¥¼ ì „ë‹¬í•œë‹¤.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()

features = iris.data
targets = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=124)

dtc = DecisionTreeClassifier()
parameters = {'max_depth': [2, 3, 4], 'min_samples_split': [6, 7]}
```

```python
g_dtc = GridSearchCV(dtc, param_grid=parameters, cv=5, refit=True, return_train_score=True, n_jobs=-1)
```

```python
g_dtc.fit(X_train, y_train)
```

```python
g_dtc.cv_results_
```

```python
result_df = pd.DataFrame(g_dtc.cv_results_)
result_df
# mean_test_score ê°€ ì œì¼ ë†’ì€ê²Œ ìµœì ì˜ íŒŒë¼ë¯¸í„°
```

```python
print(g_dtc.best_params_, g_dtc.best_score_, sep="\n")
```

```python
g_dtc.best_estimator_
```

```python
dtc = g_dtc.best_estimator_
prediction = dtc.predict(X_test)
accuracy_score(y_test, prediction)
```
