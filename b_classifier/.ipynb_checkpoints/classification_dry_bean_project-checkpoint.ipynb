### 견과류 분류

#### Feature
- **Area**: 견과류의 면적.
- **Perimeter**: 견과류의 둘레 길이.
- **MajorAxisLength**: 견과류의 가장 긴 축의 길이.
- **MinorAxisLength**: 견과류의 가장 짧은 축의 길이.
- **AspectRation**: MajorAxisLength와 MinorAxisLength의 비율.
- **Eccentricity**: 견과류의 이심률 (원에서 얼마나 벗어나 있는가를 나타내는 지표).
- **ConvexArea**: 견과류를 완전히 둘러싸는 Convex Hull의 면적.
- **EquivDiameter**: 면적이 같은 완전한 원의 지름.
- **Extent**: 견과류의 bounding box(경계 상자)에 대한 견과류 면적의 비율.
- **Solidity**: 견과류의 면적 대비 Convex Hull 면적의 비율.
- **Roundness**: 견과류의 둥근 정도.
- **Compactness**: 견과류의 밀도 또는 조밀함.
- **ShapeFactor1**: 견과류의 형태에 대한 계수 1.
- **ShapeFactor2**: 견과류의 형태에 대한 계수 2.
- **ShapeFactor3**: 견과류의 형태에 대한 계수 3.
- **ShapeFactor4**: 견과류의 형태에 대한 계수 4.

#### Target
- **Class**: 견과류의 종류 또는 분류('BARBUNYA', 'BOMBAY', 'CALI', 'DERMASON', 'HOROZ', 'SEKER', 'SIRA').

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction, average='macro')
    recall = recall_score(y_test , prediction, average='macro')
    f1 = f1_score(y_test, prediction, average='macro')
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}'.format(accuracy, precision, recall, f1))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,5))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
import numpy as np

# Classifier의 Decision Boundary를 시각화 하는 함수
def visualize_boundary(model, X, y):
    fig,ax = plt.subplots()
    
    # 학습 데이타 scatter plot으로 나타내기
    ax.scatter(X[:, 0], X[:, 1], c=y, s=25, cmap='rainbow', edgecolor='k',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()
    
    # 호출 파라미터로 들어온 training 데이타로 model 학습 . 
    model.fit(X, y)
    # meshgrid 형태인 모든 좌표값으로 예측 수행. 
    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    # contourf() 를 이용하여 class boundary 를 visualization 수행. 
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow',
                           zorder=1)
```

```python
import seaborn as sns

accuracy_scores = []
precisions = []
recalls = []
f1_scores = []

# 성능 지표를 계산하고 데이터프레임으로 변환하는 함수를 정의합니다.
def get_performance_scores(models, model_names, X_test, y_test):
    scores = []
    for model, name in zip(models, model_names):
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1]  # 확률 예측 (양성 클래스에 대한)
        
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='macro')
        recall = recall_score(y_test, y_pred, average='macro')
        f1 = f1_score(y_test, y_pred, average='macro')

        accuracy_scores.append(accuracy)
        precisions.append(precision)
        recalls.append(recall)
        f1_scores.append(f1)
        
        scores.append({
            'Model': name,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1,
        })
    return pd.DataFrame(scores)
    
# 모델 리스트와 모델 이름 리스트를 준비합니다.
models = [dtc, svc, knn, rfc]  # 모델 객체 리스트
model_names = ['DTC', 'SVC', 'KNN', 'RFC']  # 모델 이름 리스트

# 성능 점수를 계산합니다.
performance_scores = get_performance_scores(models, model_names, X_test, y_test)

# 데이터프레임을 긴 형식(long format)으로 바꿉니다.
performance_melted = performance_scores.melt(id_vars='Model', var_name='Metric', value_name='Score')

# 시각화를 진행합니다.
plt.figure(figsize=(10, 6))
sns.barplot(data=performance_melted, x='Metric', y='Score', hue='Model')
plt.title('Performance Metrics of Different Models')
plt.show()
```

```python
for i in range(4):
    print(f'{model_names[i]}의 정확도: {round(accuracy_scores[i] * 100, 4)}%, 정밀도: {round(precisions[i] * 100, 4)}%, 재현율: {round(recalls[i] * 100, 4)}%, F1 Score: {round(f1_scores[i] * 100, 4)}%')
```

```python
def get_performance_voting_scores(models, model_names, X_test, y_test):
    scores = []
    for model, name in zip(models, model_names):
        y_pred = model.predict(X_test)
        # 확률 기반 지표는 소프트 보팅에만 적용합니다.
        # if hasattr(model, "predict_proba"):
        #     y_proba = model.predict_proba(X_test)[:, 1]  # 양성 클래스에 대한 확률 예측
        #     roc_auc = roc_auc_score(y_test, y_proba)
        # else:
        #     roc_auc = 0.0  # 하드 보팅의 경우 N/A로 표시

        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='macro')
        recall = recall_score(y_test, y_pred, average='macro')
        f1 = f1_score(y_test, y_pred, average='macro')
        
        scores.append({
            'Model': name,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1,
        })
    return pd.DataFrame(scores)
```

```python
# 모델 리스트와 모델 이름 리스트를 준비합니다.
models = [s_vc, h_vc]  # 모델 객체 리스트
model_names = ['Soft Voting', 'Hard Voting']  # 모델 이름 리스트

# 성능 점수를 계산합니다.
performance_scores = get_performance_voting_scores(models, model_names, X_test, y_test)

# 데이터프레임을 긴 형식(long format)으로 바꿉니다.
performance_melted = performance_scores.melt(id_vars='Model', var_name='Metric', value_name='Score')

# 시각화를 진행합니다.
plt.figure(figsize=(10, 6))
sns.barplot(data=performance_melted, x='Model', y='Score', hue='Metric')
plt.title('Performance Metrics of Different Models')
plt.show()
```
