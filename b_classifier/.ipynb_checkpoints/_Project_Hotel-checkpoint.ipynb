```python
# Îç∞Ïù¥ÌÑ∞ Î≥µÏÇ¨
h_df = hotel_df.copy()
```

```python
h_df['booking status'].value_counts()
```

```python
# Î∂àÌïÑÏöîÌïú Ïª¨Îüº Ï†úÍ±∞
h_df.drop(labels=['Booking_ID',
                  'P-C',
                  'P-not-C',
                  'date of reservation'
                  ], axis=1, inplace=True)
display(h_df)
```

```python
h_df['type of meal'].value_counts()
```

```python
# "type of meal" Ïª¨ÎüºÏóêÏÑú "Meal Plan 1", "Meal Plan 2", "Meal Plan 3"ÏùÑ "Selected"Î°ú Î≥ÄÌôò
h_df['type of meal'] = h_df['type of meal'].replace(['Meal Plan 1', 'Meal Plan 2', 'Meal Plan 3'], 'Selected')
```

```python
h_df['type of meal'].value_counts()
```

```python
h_df['room type'].value_counts()
```

```python
h_df['room type'] = h_df['room type'].apply(lambda x: x[10])
```

```python
h_df['room type'].value_counts()
```

```python
# Î¨∏ÏûêÏó¥ Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò
from sklearn.preprocessing import LabelEncoder

meal_encoder = LabelEncoder()
meals = meal_encoder.fit_transform(h_df['type of meal'].tolist())
h_df['type of meal'] = meals

market_encoder = LabelEncoder()
markets = market_encoder.fit_transform(h_df['market segment type'].tolist())
h_df['market segment type'] = markets

booking_encoder = LabelEncoder()
bookings = booking_encoder.fit_transform(h_df['booking status'].tolist())
h_df['booking status'] = bookings

h_df
```

```python
h_df.hist(figsize=(10, 10))
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

std = StandardScaler()

features = ['number of adults', 'number of children', 'number of weekend nights',
       'number of week nights', 'type of meal', 'car parking space',
       'room type', 'lead time', 'market segment type', 'repeated',
       'average price', 'special requests', 'booking status']

total_count = h_df.shape[0]
for feature in features:
    std_df = std.fit_transform(h_df[[feature]])
    count = ((std_df < -1.96) | (std_df > 1.96)).sum()
    print(f'{feature}: {count}Í±¥, {np.round(count / total_count * 100, 4)}%')
```

```python
# Ïù¥ÏÉÅÏπò Ï†úÍ±∞
sbp_df = h_df.loc[:, 'number of adults'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['number of adults']])
sbp_df['number of adults'] = std_sbp_df
sbp_df = sbp_df[sbp_df['number of adults'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'number of children'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['number of children']])
sbp_df['number of children'] = std_sbp_df
sbp_df = sbp_df[sbp_df['number of children'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'number of weekend nights'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['number of weekend nights']])
sbp_df['number of weekend nights'] = std_sbp_df
sbp_df = sbp_df[sbp_df['number of weekend nights'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'number of week nights'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['number of week nights']])
sbp_df['number of week nights'] = std_sbp_df
sbp_df = sbp_df[sbp_df['number of week nights'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'type of meal'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['type of meal']])
sbp_df['type of meal'] = std_sbp_df
sbp_df = sbp_df[sbp_df['type of meal'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'car parking space'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['car parking space']])
sbp_df['car parking space'] = std_sbp_df
sbp_df = sbp_df[sbp_df['car parking space'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'room type'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['room type']])
sbp_df['room type'] = std_sbp_df
sbp_df = sbp_df[sbp_df['room type'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'lead time'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['lead time']])
sbp_df['lead time'] = std_sbp_df
sbp_df = sbp_df[sbp_df['lead time'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'market segment type'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['market segment type']])
sbp_df['market segment type'] = std_sbp_df
sbp_df = sbp_df[sbp_df['market segment type'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'repeated'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['repeated']])
sbp_df['repeated'] = std_sbp_df
sbp_df = sbp_df[sbp_df['repeated'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'average price'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['average price']])
sbp_df['average price'] = std_sbp_df
sbp_df = sbp_df[sbp_df['average price'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)

sbp_df = h_df.loc[:, 'special requests'].reset_index()
std_sbp_df = std.fit_transform(sbp_df[['special requests']])
sbp_df['special requests'] = std_sbp_df
sbp_df = sbp_df[sbp_df['special requests'].between(-1.96, 1.96)]
h_df = h_df.iloc[sbp_df.index].reset_index(drop=True)
```
